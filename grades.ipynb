```python
import numpy as np
import tensorflow as tf
import pandas as pd

data = pd.read_csv('gpascore.csv')
data = data.dropna()
```

```python
y_data = data['admit'].values
print(y_data)
```

```python
x_data = []
for i, rows in data.iterrows(): #한행식 출력
    x_data.append([ rows['gre'], rows['gpa'], rows['rank'] ])

print(x_data)
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64, activation = 'tanh'),
    tf.keras.layers.Dense(128, activation = 'tanh'),
    tf.keras.layers.Dense(1, activation = 'sigmoid'),
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(np.array(x_data), np.array(y_data), epochs = 20) #numpy array or tensor only
```

```python
#loss = 손실값 (실제 데이터 - 예측값)
```

```python
#예측
model.predict( [[750, 3.70, 3], [400, 2.2, 1] ] ) #그냥 x데이터를 넣는거
```
